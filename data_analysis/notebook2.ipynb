# Тестирование работы функции по получению данных

```python
from analyzer import TimerowForAnalise
from meteostat import Point, Daily
import pandas as pd
from datetime import datetime
```

```python
# Точка,в которой смотрим температуру - в данном случае Париж
paris = Point(37.77493, -122.41942)

# Временной диапазон
start = datetime(2024, 9, 1)
end = datetime(2024, 9, 10)

# Получение данных о погоде
data = Daily(paris, start, end).fetch()
```

```python
# 'tavg' - температура
temperature_data = data['tavg']
```

```python
fetcher = WeatherData([37.77493, -122.41942], '2024-09-01', '2024-09-10')
weather_data = fetcher.get_weather_data()
```

```python
combined_df = pd.concat([temperature_data, weather_data], axis=1)
```
