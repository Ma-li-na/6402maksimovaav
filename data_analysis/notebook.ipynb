```python
from analyzer import TimerowForAnalise
from meteostat import Point, Daily
import pandas as pd
from datetime import datetime
```

```python
# Точка,в которой смотрим температуру - в данном случае Париж
paris = Point(48.8566, 2.3522)

# Временной диапазон
start = datetime(2024, 9, 1)
end = datetime(2024, 10, 1)

# Получение данных о погоде
data = Daily(paris, start, end).fetch()
```

```python
# Преобразуем индексы в столбцы с датами, чтобы избежать предупреждений
data.reset_index(inplace=True)

# 'tavg' - температура
temperature_data = data['tavg'].to_numpy()

# Создаем анализатор
analyzer = TimerowForAnalise(temperature_data)
```

```python
# Вычисляем скользящее среднее
results_df = analyzer.get_results()
```

```python
# Объединяем исходные данные и результаты
results_df = pd.concat([data['time'], results_df], axis=1)
```

```python
# Сохраняем DataFrame в Excel
excel_file = 'results.xlsx'
try:
    with pd.ExcelWriter(excel_file, mode='w', engine='openpyxl') as writer:
        results_df.to_excel(writer, sheet_name='Результаты', index=False)
except FileNotFoundError:
 # Если файл не существует, создаем его
 with pd.ExcelWriter(excel_file, mode='w', engine='openpyxl') as writer:
    results_df.to_excel(writer, sheet_name='Результаты', index=False)

print(f"DataFrame сохранен в файл: {excel_file}")
```
